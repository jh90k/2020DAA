```python
import requests
from bs4 import BeautifulSoup

url='https://search.naver.com/search.naver?where=news&sm=tab_jum&query=성균관대'
raw=requests.get(url)
html=BeautifulSoup(raw.text, 'html.parser')
article=html.select('ul.type01>li')
```

```python
for i in article:
    title=i.select_one('dt>a._sp_each_title').text.strip().replace(',', '-')
    pub=i.select_one('dd.txt_inline>span._sp_each_source').text.strip().replace(',', '-')
    
    print(title)
    print(pub)
    print('='*60)
```

```python
f = open('naverarticle.csv', 'w')
f.write('기사제목, 언론사\n')

url='https://search.naver.com/search.naver?where=news&sm=tab_jum&query=성균관대'
raw=requests.get(url)
html=BeautifulSoup(raw.text, 'html.parser')
article=html.select('ul.type01>li')

for i in article:
    title=i.select_one('dt>a._sp_each_title').text.strip().replace(',','-')
    pub=i.select_one('dd>span._sp_each_source').text.strip().replace(',','-')
    
    print(title)
    print(pub)
    print('='*60)
    f.write(title+','+pub+'\n')
    
f.close()
```
